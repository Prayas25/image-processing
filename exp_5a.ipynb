```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread("/content/prayas.jpg", cv2.IMREAD_GRAYSCALE)

# Apply Sobel operator
grad_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Gradient in X direction
grad_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Gradient in Y direction

# Compute gradient magnitude
gradient_magnitude = np.sqrt(grad_x**2 + grad_y**2)
gradient_magnitude = cv2.normalize(gradient_magnitude, None, 0, 255, cv2.NORM_MINMAX)

# Convert gradients to absolute values for visualization
grad_x_abs = cv2.convertScaleAbs(grad_x)
grad_y_abs = cv2.convertScaleAbs(grad_y)
gradient_magnitude_abs = cv2.convertScaleAbs(gradient_magnitude)

# Display results
plt.figure(figsize=(12, 6))
plt.subplot(1, 4, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 4, 2)
plt.imshow(grad_x_abs, cmap='gray')
plt.title("Sobel Gradient (X)")
plt.axis("off")

plt.subplot(1, 4, 3)
plt.imshow(grad_y_abs, cmap='gray')
plt.title("Sobel Gradient (Y)")
plt.axis("off")

plt.subplot(1, 4, 4)
plt.imshow(gradient_magnitude_abs, cmap='gray')
plt.title("Gradient Magnitude")
plt.axis("off")

plt.show()
```
