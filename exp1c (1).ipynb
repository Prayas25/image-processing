```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def translate_image(image, tx, ty):
    """
    Translates the image by tx pixels in the x-direction and ty pixels in the y-direction.

    Parameters:
    image (numpy array): The input image to be translated.
    tx (int): The translation along the x-axis (horizontal shift).
    ty (int): The translation along the y-axis (vertical shift).

    Returns:
    numpy array: The translated image.
    """
    # Get the dimensions of the image
    height, width = image.shape[:2]

    # Define the translation matrix for the transformation
    translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])

    # Perform the translation using cv2.warpAffine
    translated_image = cv2.warpAffine(image, translation_matrix, (width, height))

    return translated_image

# Load an image (grayscale or color)
image = cv2.imread('/content/ganeshji.jpeg')

# Translate the image
tx, ty = 100, 50  # Example translation of 100 pixels in x and 50 pixels in y
translated_image = translate_image(image, tx, ty)

# Display the original and translated images
plt.figure(figsize=(12, 8))
plt.subplot(1, 2, 1), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image'), plt.axis('off')
plt.subplot(1, 2, 2), plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB)), plt.title('Translated Image'), plt.axis('off')
plt.show()
```
