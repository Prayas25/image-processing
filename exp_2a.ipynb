```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Mean Smoothing Filter
def mean_filter(image, kernel_size=5):
    return cv2.blur(image, (kernel_size, kernel_size))

# Gaussian Smoothing Filter
def gaussian_filter(image, kernel_size=5, sigma=0):
    return cv2.GaussianBlur(image, (kernel_size, kernel_size), sigma)

# Median Smoothing Filter
def median_filter(image, kernel_size=5):
    return cv2.medianBlur(image, kernel_size)

# Load an image (grayscale or color)
image = cv2.imread('/content/ganeshji.jpeg')

# Apply Mean Smoothing Filter
mean_smoothed_image = mean_filter(image, kernel_size=5)

# Apply Gaussian Smoothing Filter
gaussian_smoothed_image = gaussian_filter(image, kernel_size=5)

# Apply Median Smoothing Filter
median_smoothed_image = median_filter(image, kernel_size=5)

# Display the original and smoothed images
plt.figure(figsize=(12, 8))
plt.subplot(2, 3, 1), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image'), plt.axis('off')
plt.subplot(2, 3, 2), plt.imshow(cv2.cvtColor(mean_smoothed_image, cv2.COLOR_BGR2RGB)), plt.title('Mean Filtered'), plt.axis('off')
plt.subplot(2, 3, 3), plt.imshow(cv2.cvtColor(gaussian_smoothed_image, cv2.COLOR_BGR2RGB)), plt.title('Gaussian Filtered'), plt.axis('off')
plt.subplot(2, 3, 4), plt.imshow(cv2.cvtColor(median_smoothed_image, cv2.COLOR_BGR2RGB)), plt.title('Median Filtered'), plt.axis('off')
plt.show()
```
