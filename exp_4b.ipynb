```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image in grayscale
image = cv2.imread("/content/prayas.jpg", cv2.IMREAD_GRAYSCALE)

# Define Roberts cross kernels
roberts_x = np.array([[1, 0], [0, -1]], dtype=np.float32)  # Kernel for X-gradient
roberts_y = np.array([[0, 1], [-1, 0]], dtype=np.float32)  # Kernel for Y-gradient

# Apply convolution using filter2D
grad_x = cv2.filter2D(image, cv2.CV_64F, roberts_x)
grad_y = cv2.filter2D(image, cv2.CV_64F, roberts_y)

# Compute gradient magnitude
gradient_magnitude = np.sqrt(grad_x**2 + grad_y**2)
gradient_magnitude = cv2.normalize(gradient_magnitude, None, 0, 255, cv2.NORM_MINMAX)

# Display results
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(np.abs(grad_x), cmap='gray')
plt.title("Roberts Gradient (X)")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(np.abs(grad_y), cmap='gray')
plt.title("Roberts Gradient (Y)")
plt.axis("off")

plt.show()
```
