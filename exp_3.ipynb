```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def plot_histogram(image):
    """
    Plots the histogram of the input image (can handle grayscale and color images).

    Parameters:
    image (numpy array): The input image (grayscale or color).
    """
    # If the image is in color, split it into RGB channels
    if len(image.shape) == 3:
        colors = ('b', 'g', 'r')  # Blue, Green, Red channels
        plt.figure(figsize=(12, 6))
        for i, color in enumerate(colors):
            # Calculate histogram for each channel
            hist = cv2.calcHist([image], [i], None, [256], [0, 256])
            plt.plot(hist, color=color)
            plt.xlim([0, 256])
        plt.title('Color Histogram (RGB)')
    else:
        # If the image is grayscale, just calculate and plot one histogram
        hist = cv2.calcHist([image], [0], None, [256], [0, 256])
        plt.plot(hist, color='k')  # Black line for grayscale
        plt.title('Grayscale Histogram')

    # Show the histogram
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.show()

# Load an image (grayscale or color)
image = cv2.imread('/content/prayas.jpg')

# If it's a color image, convert it to RGB (OpenCV loads images in BGR by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Plot the histogram of the image
plot_histogram(image_rgb)
```
